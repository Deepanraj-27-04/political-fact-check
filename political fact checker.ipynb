```python
data=[]
for i in range(0,21152):

    review=df['statement'][i]
#a)remove un neccessary .,
    review=re.sub('[^a-zA-Z]',' ',review)
#b) lower case the text
    review=review.lower()
#c)split the text
    review=review.split()
#4.stemming
#5. remove stop words
    ps=PorterStemmer()
    review=[ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
#6.join the splitted data
    review=' '.join(review)
    data.append(review)
```

```python
data
```

#7.vectorization text to numerical format



```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf=TfidfVectorizer(max_features=20000)
```

```python
x=tf.fit_transform(data).toarray()
y=df.iloc[:,0]
```

```python
y
```

```python
x.shape
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1000)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

logreg = LogisticRegression(solver='lbfgs',max_iter=1000)

# Train the model using the training sets and check score
logreg.fit(X_train, y_train)

# Predict Output
log_predicted= logreg.predict(X_test)

logreg_score = round(logreg.score(X_train, y_train) * 100, 2)
logreg_score_test = round(logreg.score(X_test, y_test) * 100, 2)

# Equation coefficient and Intercept
print('Logistic Regression Training Score: \n', logreg_score)
print('Logistic Regression Test Score: \n', logreg_score_test)

print('Accuracy: \n', round(accuracy_score(y_test,log_predicted)*100,2))
print('Confusion Matrix: \n', confusion_matrix(y_test,log_predicted))
print('Classification Report: \n', classification_report(y_test,log_predicted))
```

```python
from sklearn.naive_bayes import MultinomialNB
mn = MultinomialNB()
mn.fit(X_train, y_train)
# Predict Output
mn_predicted = mn.predict(X_test)

mn_score = round(mn.score(X_train, y_train) * 100, 2)
mn_score_test = round(mn.score(X_test, y_test) * 100, 2)
print('Multinomial Score: \n', mn_score)
print('Multinomial Test Score: \n', mn_score_test)

print('Accuracy: \n', round(accuracy_score(y_test, mn_predicted)*100,2))
print(confusion_matrix(y_test,mn_predicted))
print(classification_report(y_test,mn_predicted))
```

```python
from sklearn.svm import SVC,LinearSVC
```

```python
from sklearn.svm import SVC,LinearSVC
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
svm_model_score = round(svm_model.score(X_train, y_train) * 100, 2)
svm_model_score_test = round(svm_model.score(X_test, y_test) * 100, 2)

print('svm_model_score: \n', svm_model_score)
print('svm_model_score_test: \n', svm_model_score_test)

Y_predict = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test,Y_predict)*100
print("Accuracy:",svm_accuracy)
cm = confusion_matrix(y_test,Y_predict)
print("Confusion Matrix is :", cm)
print(classification_report(y_test,Y_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
# Predict Output   
rf_predicted = random_forest.predict(X_test)

random_forest_score = round(random_forest.score(X_train, y_train) * 100, 2)
random_forest_score_test = round(random_forest.score(X_test, y_test) * 100, 2)
print('Random Forest Score: \n', random_forest_score)
print('Random Forest Test Score: \n', random_forest_score_test)
print('Accuracy: \n', accuracy_score(y_test,rf_predicted))
print(confusion_matrix(y_test,rf_predicted))
print(classification_report(y_test,rf_predicted))
```
